# 2D Line Plot

```python
# using pandas dataframe
df = pd.read_csv("/content/drive/MyDrive/New_datasets/sharma-kohli.csv")
df

```

```python
# plotting single plots

plt.plot(df['index'],df['V Kohli'])
```

```python
# plotting multiple plots

plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])
```

```python
# Labels title

plt.plot(df['index'],df['V Kohli'])
plt.plot(df['index'],df['RG Sharma'])

plt.xlabel('Season')
plt.ylabel('Scores')
plt.title('Career Camparison')
```

```python
# colors(hex)

plt.plot(df['index'],df['V Kohli'],color='#d77661')
plt.plot(df['index'],df['RG Sharma'],color='#d3fedb')

plt.xlabel('Season')
plt.ylabel('Scores')
plt.title('Career Camparison')
```

```python
#  line(width and style)
plt.plot(df['index'],df['V Kohli'],color='#d77661',linestyle='dotted',linewidth=3)
plt.plot(df['index'],df['RG Sharma'],color='#d3fedb',linestyle='dashdot',linewidth=2)

plt.xlabel('Season')
plt.ylabel('Scores')
plt.title('Career Camparison')
```

```python
# marker(size)

plt.plot(df['index'],df['V Kohli'],color='#d77661',linestyle='dotted',linewidth=3,marker='o',markersize=5)
plt.plot(df['index'],df['RG Sharma'],color='#d3fedb',linestyle='dashdot',linewidth=2,marker='D',markersize=8)

plt.xlabel('Season')
plt.ylabel('Scores')
plt.title('Career Camparison')
```

```python
# legentd > location

plt.plot(df['index'],df['V Kohli'],color='#d77661',linestyle='dotted',linewidth=3,marker='o',markersize=5,label='Virat')
plt.plot(df['index'],df['RG Sharma'],color='#d3fedb',linestyle='dashdot',linewidth=2,marker='D',markersize=8,label='Kohli')

plt.xlabel('Season')
plt.ylabel('Scores')
plt.title('Career Camparison')
plt.legend()
```

```python
# limiting axes
# trimming values when an outlier comes in your graph

price = [48000,54000,57000,49000,47000,45000,450000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,100000)
plt.xlim(2017,2019)
```

```python
# grid view

plt.plot(df['index'],df['V Kohli'],color='#d77661',linestyle='dotted',linewidth=3,marker='o',markersize=5,label='Virat')
plt.plot(df['index'],df['RG Sharma'],color='#d3fedb',linestyle='dashdot',linewidth=2,marker='D',markersize=8,label='Kohli')

plt.xlabel('Season')
plt.ylabel('Scores')
plt.title('Career Camparison')
plt.legend()
plt.grid()
```

```python
# show

plt.plot(df['index'],df['V Kohli'],color='#d77661',linestyle='dotted',linewidth=3,marker='o',markersize=5,label='Virat')
plt.plot(df['index'],df['RG Sharma'],color='#d3fedb',linestyle='dashdot',linewidth=2,marker='D',markersize=8,label='Kohli')

plt.xlabel('Season')
plt.ylabel('Scores')
plt.title('Career Camparison')
plt.legend()
plt.grid()
plt.show() # main function which plots graphs
```

# **Bar Chart**

```python
# simple bar chart

children = [10,20,30,40,50]
colors = ['red','green','blue','pink','gray']

plt.bar(colors,children)
```

```python
# horizontal bar chart

plt.barh(colors,children,color='red')
```

```python
# color and label

df = pd.read_csv('/content/drive/MyDrive/New_datasets/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
# multiple bar charts

plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
# a problem
children = [10,20,30,40,50]
colors = ['red red red red','green green green green','blue blue','pink pink','gray gray']

plt.bar(colors,children)
plt.xticks(rotation='vertical')
```

# **Stacked Bar chart**

```python
# Stacked bard chart

plt.bar(df['batsman'],df['2015'],label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2015']) + df['2016'],label='2017')

plt.legend()
plt.show()
```

# **Pie Chart**

*   Univariate/Bivariate Analysis
*   Categorical vs Numerical
*   Usecase - To find contribution on a standard scale





```python
# simple data

data = [23,25,100,49,20]
subjects = ['science','math','coding','machine learning','statistics']

plt.pie(data,labels=subjects)
plt.show()
```

```python
# dataset

df = pd.read_csv('/content/drive/MyDrive/New_datasets/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and colors

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['red','blue','green','black','cyan'])
plt.show()
```

```python
# explode shadow

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0.1],shadow=True)
plt.show()
```
